##### -_**RENAME THE (NaN) values in [mes_nombre] based on the values in [mes]**_-
- _Convert the values in **[mes]** to int._

```python
# Define a dictionary to map each ("month number") to it's name
mes_dict =  {1: 'enero', 
             2: 'febrero', 
             3: 'marzo', 
             4: 'abril', 
             5: 'mayo',
             6: 'junio',
             7: 'julio', 
             8: 'agosto', 
             9: 'septiembre', 
             10: 'octubre',
             11: 'noviembre', 
             12: 'diciembre'}

#Fill the ("NaN") values in _[mes_nombre]_ based on _[mes]_ using the dictionary ("mes_dict")
df['mes_nombre'] = df.apply(
    lambda row: mes_dict[int(row['mes'])] if pd.isna(row['mes_nombre']) and not pd.isna(row['mes']) else row['mes_nombre'],
    axis=1
)
df = df.dropna()
df
```

```python
# Convert ['mes'] from float to integer
df['mes'] = df['mes'].astype(int)
```

```python
# Convert ['periodo'] from float to integer
df['periodo'] = df['periodo'].astype(int)
```

```python
df['dia'] = df['dia'].astype(int)
```

```python
df.set_index('OBJECTID', inplace=True)
```

### -_**FINALLY WE'RE GOING THROUGH THE [DATE] DATA TYPE**_-
- _Convert this, ex: __[5/5/2017 0:00]__ to **[5-5-2017]**_
- _Convert our **[hora]** column from (**standard**) hour to **military hour**_

```python
# Convert 'fecha' to datetime format, handling the initial format in your data
df['fecha'] = pd.to_datetime(df['fecha'], errors='coerce')

# Format the 'fecha' column to the desired format (day-month-year) and remove the time part
df['fecha'] = df['fecha'].dt.strftime('%d-%m-%Y')

# Preview the updated 'fecha' column
print(df['fecha'].head())
```

```python
df
```

```python
# Convert 'hora' to datetime format to parse the time correctly
df['hora'] = pd.to_datetime(df['hora'], format='%I:%M %p', errors='coerce')

# Format 'hora' to 24-hour military format (HH:MM:SS)
df['hora'] = df['hora'].dt.strftime('%H:%M:%S')

# Preview the updated 'hora' column
print(df['hora'].head())
```
